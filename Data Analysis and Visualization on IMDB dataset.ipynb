### Histogram

### Box Plot 
1) It is used to represent the quartile/percentile distribution of data<br>
2) It is used to visualize distribution of numerical data<br>
3) It represnets a 5 point summary<br>
    
    Q1,Q2,Q3,min and max

i) <b>Q1</b> : 25 percentile <br>
ii) <b>Q2</b> :  50 percentile (median)<br>
iii) <b>Q3</b> :  75 percentile <br>
iv) <b>IQR (Inter Quartile Range)</b> : Q3-Q1<br>
v) <b>min</b> : = Q1 - 1.5 IQR<br>
vi) <b>max</b> : = Q3 + 1.5 IQR<br>
vii) <b>Outliers</b> : >max or <min

#### Demo Example

```python
x = [12,14, 16, 15, 14,25, 12, 11, 9,10,13,15,2,17,9,16]
print(x)
print(len(x))
```

```python
plt.boxplot(x)
plt.grid()
plt.show()
# q3 - 15.25
# q2 - 13.5
# q1 - 10.75 
# max - 22 (actual)
# min - 4 (actual)
# max in the graph = value closest to 22 less than outlier 25 in the list x = 17
# min in the graph = value closest to 2 grater than outlier 2 in the list x = 9
```

```python
q1 = np.percentile(x,25)
q2 = np.percentile(x,50) # Median
q3 = np.percentile(x,75)
iqr = q3-q1    # Inter Quartile Range
print('Q1',q1)
print('Q2',q2)  # Median
print('Q3',q3)
print('IQR',iqr)
bmax = q3 + 1.5*iqr 
bmin = q1 - 1.5*iqr
print('Max',bmax)
print('Min',bmin)
x.sort()
print(x)
```

### Stack Bar Chart

Multiple column data values when represented into bars are placed on top of another bar is referred to as Stacked Bar Chart

#### Q) Plot mean of Budget and Profit based on genre1 on a Stacked Bar Chart

```python
r1 = d1.groupby('genre_1')[['budget','Profit']].mean()
r1.index.names = ['Genre']
r1
```

#### Overlapped values

```python
plt.bar(r1.index,r1['budget'],color='cyan',label='Mean Budget')
plt.bar(r1.index,r1['Profit'],color='orange',label='Mean Profit')
plt.xlabel('Genre')
plt.legend()
plt.title('Genre wise mean Budget and Profit')
plt.xticks(rotation=90)
plt.grid()
plt.ylim(0,280)
plt.show()
```

#### Using the bottom parameter to create stacked bar charts

```python
plt.bar(r1.index,r1['budget'],color='cyan',label='Mean Budget')
plt.bar(r1.index,r1['Profit'],color='orange',label='Mean Profit',bottom=r1['Profit'])
plt.xlabel('Genre')
plt.legend()
plt.title('Genre wise mean Budget and Profit')
plt.xticks(rotation=90)
plt.grid()
plt.ylim(0,280)
plt.show()
```

### Multiple Plots
