```python
# Importing Matplotlib
import matplotlib.pyplot as plt
# plt is athe alias name for pyplot
import numpy as np
# np is the alias for numpy
import pandas as pd
# pd is the alias for pandas
import seaborn as sns
# seaborn is aliased as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading Dataset
df = pd.read_csv("D:/TESTBOOK DATA SCIENCE WITH PYTHON/Advance python code file/Python - Notebooks/tips.csv")
# Print the Type of df
print(type(df))
# Show top 5 rows
df.head()  
```

```python
df.count() # count the total no of entries in the dataframe 
```

```python
df['day'].unique() # finding unique values in day column
```

```python
# Making a new dataframe by grouping the data according to the day column
# Calculating the mean of numeric columns according to group by condition and resetting the index
df_day_mean = df.groupby('day').mean().reset_index()
 # Rounding 'total_bill','tip','size' values
df_day_mean[['total_bill','tip','size']]=round(df_day_mean[['total_bill','tip','size']])
# Converting the float values in 'total_bill','tip','size' to int values
df_day_mean[['total_bill','tip','size']] = df_day_mean[['total_bill','tip','size']].astype(int) 
# Showing the dataframe
df_day_mean
```

```python
df_day_mean.dtypes # Finding the datatypes of all columns
```

Plotting a normal bar plot of 'Day vs  Mean Total bill'

```python
# Method - 1
sns.barplot(x=df_day_mean['day'],y=df_day_mean['total_bill'])
```

```python
# Method - 2
sns.barplot(data=df_day_mean,x='day',y='total_bill')
# Showing the plot
plt.show()
```

Styling Bar Plot

```python
# Plotting  the bar plot with color = 'teal' and ci = None
sns.barplot(data=df_day_mean,x ='day',y='size',ci=None,color='teal')
# Setting title for the plot
plt.title('Mean of Size based on day')
# Showing the plot
plt.show()
```

```python
# Making a new dataframe by grouping the data according to the 'day' and 'time' column
# Calculating the sum of numeric columns according to group by condition and resetting the index
df_day_time = df.groupby(['day','time']).sum().reset_index()
# Converting the float values in 'total_bill','tip','size' to int values
df_day_time[['total_bill','tip','size']] = df_day_time[['total_bill','tip','size']].astype(int) 
# Showing the dataframe
df_day_time
```

```python
# Plotting  the bar plot with hue ='time' and ci = None
sns.barplot(data=df_day_time, x='day',y= 'tip',hue='time',ci=None)
# Setting title for the plot
plt.title('Sum of Tips based on Day and Time')
# Showing the grid for the plot
plt.grid()
# Showing the plot
plt.show()
```

Plotting a simple line plot

```python
# Plotting  the line  plot with ci = None
sns.lineplot(data=df_day_mean, x='day',y= 'total_bill',ci=None)
# Showing the plot
plt.show()
```

```python
# Making a new dataframe by grouping the data according to the 'day','time' and sex column
# Calculating the sum of numeric columns according to group by condition and resetting the index
df_day_sum = df.groupby(['day','time','sex']).sum().reset_index()
# Converting the float values in 'total_bill','tip','size' to int values
df_day_sum[['total_bill','tip','size']] = df_day_sum[['total_bill','tip','size']].astype(int) 
# Showing the dataframe
df_day_sum
```

```python
# Plotting  the line  plot with hue ='sex' and ci = None
sns.lineplot(data=df_day_sum, x='day',y= 'tip',hue='sex',ci=None)
# Setting title for the plot
plt.title('Sum of Tips based on Day and Sex')
# Showing the plot
plt.show()
```

```python
# Plotting  the line  plot with hue ='time', style ='time' and markers = True
sns.lineplot(data=df_day_sum, x='day',y= 'tip',hue='sex',style='sex',markers=True)
# Setting title for the plot
plt.title('Sum of Tips based on Day and Sex')
# Showing the plot
plt.show()
```

# Scatter plot

```python
# Plotting a scatter plot
sns.scatterplot(data=df,x='total_bill',y='tip')
# Setting title for the plot
plt.title('Total Bill vs Tips')
plt.show() # Showing the plot
```

```python
plt.figure(figsize=(8,6)) # setting plot figure size
# Plotting a scatter plot and applying customization
sns.scatterplot(data=df,x='total_bill',y='tip',palette='Set1',hue="day", style="time")
# Setting title for the plot
plt.title('Total Bill vs Tips based on Day')
plt.legend(loc=9) # Setting Location of legend inside the plot
plt.grid() # Showing grid for the plot 
plt.show()# Showing the plot
```

```python
plt.figure(figsize=(8,4)) # setting plot figure size
# Plotting a scatter plot and applying customization
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex',
                palette='Set1',size='sex')
# Setting title for the plot
plt.title('Total Bill vs Tips based on Sex')
# Setting position and anchoring the legend outside the plot
plt.legend(bbox_to_anchor=(1, 1))
plt.grid() # Showing grid for the plot 
# Showing the plot
plt.show()
```
