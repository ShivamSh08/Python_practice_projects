### 1. How has honey production yield changed from 1998 to 2021?

```python
a0 = df.groupby('year')['yieldpercol'].sum().reset_index()
a0
```

```python
plt.bar(a0['year'], a0['yieldpercol'],color=['red','skyblue','violet','orange','yellow','blue','cyan','brown'])
plt.title('Year Wise Total Of Yield Per Colonies')
plt.legend()
plt.xlabel('For Year1998 - 2021')
plt.xticks(rotation=90)
plt.ylabel('Yield per colonies')
plt.show()
```

```python
a1 = df.groupby('year')['yieldpercol'].mean().reset_index()
a1
```

```python
plt.figure(figsize=(8,4))
plt.plot(a1['year'], a1['yieldpercol'],marker ='o', markersize = '10',linestyle='-.')
plt.title('Year Wise Average Of Yield Per Colonies')
plt.xlabel('For Year1998 - 2021')
plt.xticks(rotation=90)
plt.ylabel('Yield per colonies')
plt.grid()
plt.show()
```

#### Observations: <br>
It seems like that there is a overall downward trend in yield per colony.The trend is showing there is declined in yield per colony.

### 3. Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006, which was when concern over Colony Collapse Disorder spread nationwide?

```python
a2  = df[df['year']<=2006].groupby('year')[['numcol','yieldpercol']].sum()
a2
```

```python
a3  = df[df['year']<=2006].groupby('year')[['numcol','yieldpercol']].mean()
a3
```

```python
plt.figure(figsize=(12,8))
plt.plot(a2.index,a2['numcol'],color='red',marker='D')
plt.title('Total Number of Colonies Year Wise')
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Total Number of colonies', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(a3.index,a3['numcol'],color='red',marker='D')
plt.title('Average Number of Colonies Year Wise', fontsize = 15)
plt.xlabel('Year',fontsize = 15)
plt.ylabel('Total Number of colonies',fontsize = 15)
plt.show()
```

### Observation: <br>
As per 'Total number of colonies' and "Average number of colonies" year wise
we can cleary see that there is a sudden drop in number of colonies from year 2004 to 2006, which have raised a concern

```python
plt.figure(figsize=(12,8))
plt.plot(a2.index,a2['yieldpercol'],color='red',marker='D')
plt.title('Total Yield per Colony Year Wise', fontsize =15)
plt.xlabel('Year', fontsize =15)
plt.ylabel('Total Yield per Colony', fontsize =15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(a3.index,a3['yieldpercol'],color='red',marker='D')
plt.title('Average Yield per Colony Year Wise', fontsize =15)
plt.xlabel('Year', fontsize =15)
plt.ylabel('Total Number of colonies', fontsize =15)
plt.show()
```

#### Observation: <br>

As per plots that shows visualization of "Total Of Yield per colony" and "Average Yield per Colony" year wise **Yield per colony has shows fall which has been started from year 2003.**

### 4. Are there any patterns that can be observed between total honey production and value of production every year?

```python
a4 = df.groupby('year')[['totalprod','prodvalue']].sum()
a4
```

```python
a5 = df.groupby('year')[['totalprod','prodvalue']].mean()
a5
```

```python
plt.figure(figsize =(8,4))
plt.stackplot(a4.index,a4['totalprod'],a4['prodvalue'],
                          labels=['Total Production', 'Value Production'])
plt.title('Year Wise Total of TOTAL PRODUCTION AND VALUE OF PRODUCTION', fontsize=10)
plt.xlabel('Year')
plt.legend()
plt.show()
```

```python
plt.figure(figsize =(8,4))
plt.stackplot(a5.index,a5['totalprod'],a5['prodvalue'],
                          labels=['Total Production', 'Value Production'])
plt.title('Year Wise Average of TOTAL PRODUCTION AND VALUE OF PRODUCTION', fontsize=9)
plt.xlabel('Year')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(a4.index,a4['totalprod'], label='Total Production', color = 'red')
plt.bar(a4.index,a4['prodvalue'], label='Value of Production', color = 'green', alpha = 0.5)
plt.xlabel('Year')
plt.title('Year Wise Total of TOTAL PRODUCTION AND VALUE OF PRODUCTION', fontsize=9)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(a5.index,a5['totalprod'], label='Total Production', color = 'red')
plt.bar(a5.index,a5['prodvalue'], label='Value of Production', color = 'green', alpha = 0.5)
plt.xlabel('Year')
plt.title('Year Wise Average of TOTAL PRODUCTION AND VALUE OF PRODUCTION', fontsize=9)
plt.legend()
plt.show()
```

#### Observation :<br>

As per the data the Total production has been fallen while the value of Production has been has been increased.
